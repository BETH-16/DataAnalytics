```python
import numpy as np
```

## Array Basics

```python
fruits = ['apple', 'banana', 'cherry', 'date']
```

```python
arr_1 = np.array(fruits)
print(arr_1)
```

```python
print(arr_1.dtype)
# U = Unicode string
# 6 = max number of characters in the strings
# cherry has 6 characters so dtype is U6
```

```python
arr_2 = np.array([(10, 11, 12), (13, 14, 15)])
print(arr_2)
```

```python
print(arr_2.T)
```

```python
print(arr_2)
# No! .T returns a transposed view but does not change the original arr_2
```

```python
arr_3 = np.array([(1, 3, 5), (2, 4, 6)])
print(arr_3)
```

```python
# arr_3[0, 2] = row 0, column 2 = value 5
# arr_3[1, 1] = row 1, column 1 = value 4
arr_4 = np.array([arr_3[0, 2], arr_3[1, 1]])
print(arr_4)
```

```python
arr_5 = np.ones((2, 2, 2))
print(arr_5)
```

# Exercise 2.C Lab 3 Working with payroll tax functions in Python.

```python
# a) get_soc_sec_tax()
# Accepts gross pay and returns Social Security tax at 6.2%
def get_soc_sec_tax(gross_pay):
    return gross_pay * 0.062
```

```python
# b) get_medicare_tax()
# Accepts gross pay and returns Medicare tax at 1.45%
def get_medicare_tax(gross_pay):
    return gross_pay * 0.0145
```

```python
# c) get_federal_tax()
# Accepts gross pay and withholding code, returns federal tax
# Withholding codes:
# 0 = 23%, 1 = 21%, 2 = 19.5%, 3 = 18.5%
# 4+ = 18% minus 0.5% for each withholding over 4
def get_federal_tax(gross_pay, withholding_code):
    if withholding_code == 0:
        tax_rate = 0.23
    elif withholding_code == 1:
        tax_rate = 0.21
    elif withholding_code == 2:
        tax_rate = 0.195
    elif withholding_code == 3:
        tax_rate = 0.185
    else:
        # code 4 and above: 18% minus 0.5% for each withholding over 4
        extra = withholding_code - 4
        tax_rate = 0.18 - (extra * 0.005)
    return gross_pay * tax_rate
```

```python
# Test Person 1: gross pay $750, withholding code 0
gross_pay = 750
code = 0

soc_sec  = get_soc_sec_tax(gross_pay)
medicare = get_medicare_tax(gross_pay)
federal  = get_federal_tax(gross_pay, code)
total    = soc_sec + medicare + federal

print(f"Person 1 — Gross Pay: ${gross_pay:.2f}, Withholding Code: {code}")
print(f"Social Security Tax: ${soc_sec:.2f}")
print(f"Medicare Tax:        ${medicare:.2f}")
print(f"Federal Tax:         ${federal:.2f}")
print(f"Total Tax Withheld:  ${total:.2f}")
```

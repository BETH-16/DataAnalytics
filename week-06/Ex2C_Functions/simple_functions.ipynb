Lab 1 of Exercise 2.C, working with defining custom functions

```python
def favorite_things( name, movie, music):
  return  f"My name is {name}.\nMy favorite movie is {movie}.\nMy favorite musician is {music}."
```

```python
print(favorite_things('Betty', 'Green Book', 'Meghan Trainor'))
```

```python
def why_im_here(purpose):
    return purpose
```

```python
print(why_im_here('I joined Year Up United Data Analyst program to build a career in tech, gain real world data skills, and secure a full time job.'))
```

```python
def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}."
```

```python
print(favorite_place('Greece', 'it has stunning views, rich history, and delicious food'))
```

```python
print(favorite_place('Cancun', 'it has beautiful beaches, crystal clear water, and amazing weather'))
```

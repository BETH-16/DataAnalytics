# Exercise 2.A Using built-in functions

## Lab 2. Parsing strings – part codes

```python
part_code = 'ACME:123-L'
```

```python
def get_supplier_code(code):
    colon_index = code.index(':')
    return code[:colon_index]
```

```python
print(get_supplier_code(part_code))
```

```python
def get_product_num(code):
    colon_index = code.index(':')
    dash_index = code.index('-')
    return code[colon_index+1:dash_index]
```

```python
print(get_product_num(part_code))
```

```python
def get_size(code):
    dash_index = code.index('-')
    return code[dash_index+1:]
```

```python
print(get_size(part_code))
```

```python
part_code_1 = 'ACME:123-L'
part_code_2 = 'DI:12345-M'
part_code_3 = 'ACE:1-12'
```

```python
print(f"For {part_code_1}, the supplier is {get_supplier_code(part_code_1)}. Part number {get_product_num(part_code_1)} is available in size {get_size(part_code_1)}.")
```

```python
print(f"For {part_code_2}, the supplier is {get_supplier_code(part_code_2)}. Part number {get_product_num(part_code_2)} is available in size {get_size(part_code_2)}.")
```

```python
print(f"For {part_code_3}, the supplier is {get_supplier_code(part_code_3)}. Part number {get_product_num(part_code_3)} is available in size {get_size(part_code_3)}.")
```

```python
part_codes_list = ['ACME:123-L', 'DI:12345-M', 'ACE:1-12']

for code in part_codes_list:
    print(f"For {code}, the supplier is {get_supplier_code(code)}. Part number {get_product_num(code)} is available in size {get_size(code)}.")
```
